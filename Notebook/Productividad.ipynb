```python
import pandas as pd
import numpy as np
```

```python
dict_gral = pd.read_excel('../Data/horarios MADRE.xlsx', sheet_name=None)
```

```python
print(dict_gral)
```

```python
# Renombrar todas las columnas en tu diccionario
for nombre, df in dict_gral.items():
    if df is not None and not df.empty:
        df.columns = [f'col{i+1}' for i in range(len(df.columns))]
```

```python
# Une todos los DataFrames de un diccionario en uno solo

df_unificado = pd.concat(dict_gral.values(), ignore_index=True)
```

```python
df_unificado.sample(25)
```

```python
#  Une los DataFrames agregando una columna con el nombre del empleado

dfs_con_empleado = []
    
for empleado, df in dict_gral.items():
    df_temp = df.copy()
    df_temp['EMPLEADO_KEY'] = empleado
    dfs_con_empleado.append(df_temp)

df_unificado = pd.concat(dfs_con_empleado, ignore_index=True)
```
