```python
df = pd.read_excel('../Data/horarios_unificado.xlsx')
df.info()
```

```python
df.head(10)
```

```python
df.dropna(subset=['col3'], inplace=True)
```

```python
# Usar errors='coerce' - convierte valores problemticos en NaT (Not a Time)
df['col3_datetime'] = pd.to_datetime(df['col3'], errors='coerce')
df['fecha'] = df['col3_datetime'].dt.date
df['hora'] = df['col3_datetime'].dt.time
```

```python
df.sample(10)
```

```python
df['fechadt'] = pd.to_datetime(df['fecha'])
df['anio'] = df['fechadt'].dt.year
df['mes'] = df['fechadt'].dt.month
```

```python
df.sample(10)
```

```python
# Suma de horas
df['suma_col6_col8'] = df['col6'] + df['col8']
df.sample(10)
```
